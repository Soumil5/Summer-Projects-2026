```python
def evaluate_values_for_current_policy(transition_probabilities, policy, gamma, n_states, reward):
	'''
    Writing bellman equation for all states in matrix form for a deterministic policy
    Values(64x1) = (rewards*transition (dot porduct along inner most aixs))[][policy_action] (64x1) +  gamma*(transition[][policy_action](64x64) * values (64x1))
    Therefore, (I - gamma*(transition[][policy_action])* (Vaules) = (rewards*transition (dot porduct along inner most aixs))[][policy_action]
    Solve the above system of linear equations (recursive equations in actual)
    '''
    
	expected_reward_each_state= np.sum(transition_probabilities * reward, axis=2)#Dot product next state axis [64x4]
	rows_for_coordinate_matching = np.arange(n_states) #[0,1,2..64]
	
	policy_transition = transition_probabilities[rows_for_coordinate_matching, policy] #Chooses probabilities for policy action [64x64]
	policy_reward = expected_reward_each_state[rows_for_coordinate_matching, policy] #Chosses Rewards for policy action [64x1]
	
	lhs = np.eye(n_states) - gamma * policy_transition # I - gamma*probabilities
	#print(lhs)
	values = np.linalg.solve(lhs, policy_reward) # Policy reward is rhs
	return values

'''This is required only in policy iteration when multiple sweeps are required to find out correct state values or the recursive equations need to be solved exactly'''
```

```python
def plot_policy_on_frozen_lake(env, policy, title="FrozenLake policy"):
	desc = np.asarray(env.unwrapped.desc, dtype=str)
	policy_grid = np.asarray(policy).reshape(desc.shape)
	arrows = np.array(["<", "v", ">", "^"])
	colors = {
		"S": "#9be7a1",
		"F": "#dceefb",
		"H": "#3a3a3a",
		"G": "#ffd54f",
	}

	fig, ax = plt.subplots(figsize=(8, 8))
	for r in range(desc.shape[0]):
		for c in range(desc.shape[1]):
			tile = desc[r, c]
			rect = plt.Rectangle((c, desc.shape[0] - 1 - r), 1, 1, facecolor=colors[tile], edgecolor="black", linewidth=1.5)
			ax.add_patch(rect)

			if tile == "H":
				label = "H"
			elif tile == "G":
				label = "G"
			elif tile == "S":
				label = f"S{arrows[policy_grid[r, c]]}"
			else:
				label = arrows[policy_grid[r, c]]

			ax.text(c + 0.5, desc.shape[0] - 1 - r + 0.5, label, ha="center", va="center", fontsize=16, fontweight="bold", color="black")

	ax.set_xlim(0, desc.shape[1])
	ax.set_ylim(0, desc.shape[0])
	ax.set_xticks(np.arange(desc.shape[1] + 1))
	ax.set_yticks(np.arange(desc.shape[0] + 1))
	ax.grid(True, color="black", linewidth=1.0)
	ax.set_xticklabels([])
	ax.set_yticklabels([])
	ax.set_aspect("equal")
	ax.set_title(title)
	plt.tight_layout()
	plt.show()
```

```python
def policy_iteration(n_states, transition_probabilities, reward, gamma=0.001, max_iterations=1000):
	#Value of gamma doesn't matter in this as rewards only at goal state at which game ends
	policy = np.zeros(n_states, dtype=int) # Initialising policy to [0,0,..0] or all Left

	for iteration in range(max_iterations):
		values = evaluate_values_for_current_policy(transition_probabilities, policy, gamma, n_states, reward)
		future = np.einsum("ijk , k -> ij", transition_probabilities, values) #Dot product on axis 2
		expected_reward_each_state= np.sum(transition_probabilities * reward, axis=2)
		q_values= expected_reward_each_state + future*gamma #(64x4)
		new_policy = np.argmax(q_values, axis=1) 
		if np.array_equal(new_policy, policy):
			return policy
		
		else:
			policy = new_policy
			
	values = evaluate_values_for_current_policy(transition_probabilities, policy, gamma, n_states, reward)
	return policy

policy = policy_iteration(n_states, transition_probabilities_statewise_actionwise, rewards_state_wise_actionwise)
plot_policy_on_frozen_lake(env, policy)
```

```python
def run_one_sweep_to_update_values(prev_values, n_states, policy, expected_reward, transition_probabilities, gamma):
    new_values=np.zeros(n_states)
    for state in range (n_states):
        	new_values[state] = expected_reward[state][policy[state]] + gamma * np.sum(transition_probabilities[state][policy[state]] * prev_values)
    return new_values
```

```python
def value_iteration(n_states, transition_probabilities, reward, gamma=0.001, max_iterations=100000):
    policy = np.zeros(n_states, dtype=int) # Initialising policy to [0,0,..0] or all Left
    expected_reward_each_state= np.sum(transition_probabilities * reward, axis=2)#Dot product next state axis [64x4]
    prev_values = np.zeros(n_states)
    for iteration in range(max_iterations):
        new_values = run_one_sweep_to_update_values(prev_values, n_states, policy, expected_reward_each_state, transition_probabilities, gamma)
        future = np.einsum("ijk , k -> ij", transition_probabilities, new_values) #Dot product on axis 2
        
        q_values= expected_reward_each_state + future*gamma #(64x4)
        new_policy = np.argmax(q_values, axis=1)
        
        if np.array_equal(new_policy, policy):
            return policy
        else:
            prev_values = new_values
            policy = new_policy
    new_values = run_one_sweep_to_update_values(prev_values, n_states, policy, expected_reward_each_state, transition_probabilities, gamma)
    return policy

policy = value_iteration(n_states, transition_probabilities_statewise_actionwise, rewards_state_wise_actionwise)
plot_policy_on_frozen_lake(env, policy)
```
